```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import csv
import os
```

```python
file = os.path.join ('data', 'vgsales.csv')
df = pd.read_csv(file)
```

```python
df.head()
```

```python
df.Year = df.Year.fillna(0.0).astype(int)
```

```python
df = df.sort_values(by=['Year'], ascending=False)
df.head(10)
```

```python
df.sort_values(by=['Global_Sales'], ascending=False).head(1)
```

```python
df.sort_values(by=['JP_Sales'], ascending=False).head(1)
```

```python
df.sort_values(by=['Year'], ascending=True).head()
```

```python
df2 = df.sort_values(by=['Year'], ascending=True)[df.Year != 0]
df2.head()
```

```python
vg_Global = list(df2.Global_Sales)
vg_NA = list(df2.NA_Sales)
pt.plot(vg_Global)
pt.plot(vg_NA)
pt.show()
```
